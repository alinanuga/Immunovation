```python
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/Dataset.xlsx'
```

```python
data = pd.read_excel(file_path)
```

```python
data.head(10)
```

```python
data.tail()
```

```python
data.info()
```

```python
data['IC50'] = data['IC50'].replace('[^\d.]', '', regex=True).astype(float)
```

```python
data['EC50'] = data['EC50'].str.replace('[^\d.]', '', regex=True)
```

```python
data['EC50'] = data['EC50'].astype(float)
```

```python
data['Binding Affinity (KD)'] = pd.to_numeric(data['Binding Affinity (KD)'], errors='coerce')
```

```python
data['PMID'] = data['PMID'].str.replace(r'\D', '', regex=True)
```

```python
data['PMID'] = pd.to_numeric(data['PMID'], errors='coerce')
```

```python
data['Interface residues'] = data['Interface residues'].replace('()', np.nan)
```

```python
unique_values = data['Variant subtypes'].unique()
print(f"Количество уникальных значений в столбце 'Variant subtypes': {unique_values}")
```

```python
nd_count = data['VH sequence'].value_counts().get('ND', 0)
print(f"Количество значений 'ND' в столбце 'VH sequence': {nd_count}")
nd_count = data['VL sequence'].value_counts().get('ND', 0)
print(f"Количество значений 'ND' в столбце 'VL sequence': {nd_count}")
```
